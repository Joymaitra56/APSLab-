# Travelling Salesman Problem using Dynamic Programming.
Problem Description:
Given a set of cities and distances, find the shortest possible route that visits each city once and returns to the starting city.

#Theory
TSP uses dynamic programming with bit masking. Store results of visited subsets to reduce repeated work.

## Algorithm
1. Start from source city
2. Mark visited cities using bit mask
3. Try all unvisited cities
4. Choose minimum cost path
5. Return to source city

```python
def tsp(graph):
    n = len(graph)
    dp = [[-1]*(1<<n) for _ in range(n)]
    # Store the next city in the optimal path for a given (city, mask) state
    path_tracker = [[-1]*(1<<n) for _ in range(n)]

    def visit(city, mask):
        if mask == (1<<n) - 1:
            return graph[city][0] # Return to the starting city

        if dp[city][mask] != -1:
            return dp[city][mask]

        ans = float('inf')
        best_next_city = -1

        for next_city in range(n):
            # Check if next_city is not visited yet
            if not (mask & (1<<next_city)):
                new_cost = graph[city][next_city] + visit(next_city, mask | (1<<next_city))
                if new_cost < ans:
                    ans = new_cost
                    best_next_city = next_city

        dp[city][mask] = ans
        path_tracker[city][mask] = best_next_city
        return ans

    min_cost = visit(0, 1) # Start from city 0 with mask 1 (only city 0 visited)

    # Reconstruct the path
    current_path = [0] # Start from city 0
    current_mask = 1
    current_city = 0

    while len(current_path) < n:
        next_city = path_tracker[current_city][current_mask]
        current_path.append(next_city)
        current_mask |= (1 << next_city)
        current_city = next_city
    current_path.append(0) # Return to the starting city

    return min_cost, current_path
```

```python
graph = [
[0, 10, 15, 20],
[10, 0, 35, 25],
[15, 35, 0, 30],
[20, 25, 30, 0]
]

min_cost, path = tsp(graph)
print("Minimum travelling cost:", min_cost)
print("Path:", ' -> '.join(map(str, path)))
```
